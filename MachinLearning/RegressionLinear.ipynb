```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset= pd.read_csv('avito_car_dataset_ALL.csv',sep=",", encoding='latin-1')
dataset
```

```python
datafram = dataset.iloc[:,2:]
datafram
```

```python
datafram.etat.unique()
```

```python
datafram = datafram.dropna()
dataframDC = pd.get_dummies(datafram)
dataframDC
```

```python
X = dataframDC.drop("Prix", axis='columns')
Y = dataframDC.Prix
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1)
```

```python
regressor  = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
#testCar= X_test.iloc[:,:]
Car = 99
Car_features= X_test.iloc[[Car]]
Car_features
```

```python
Predict_price_car = regressor.predict(Car_features)
Predict_price_car
```

```python
Real_price_car= Y_test.iloc[[Car]]
Real_price_car
```

```python
Erreur_prediction = abs(Real_price_car-Predict_price_car)
Erreur_prediction
```
